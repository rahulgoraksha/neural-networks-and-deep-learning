```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("csv/placement_data.csv")
df
```

```python
plt.Figure(figsize=(4,3))
sns.scatterplot(x = 'CGPA', y = 'Score', data=df, hue='Placed')
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['Placed']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=7, stratify=y)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha=1)
pr.fit(X_train,y_train)
```

to verify wether model is undefit / overfit / realfit

```python
pr.score(X_train,y_train)*100, pr.score(X_test, y_test)*100 
```

> result into underfit

```python
plt.Figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = pr)
plt.show()
```
