```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/sharmaroshan/Churn-Modelling-Dataset/master/Churn_Modelling.csv"

df = pd.read_csv(url)
df
```

```python
df.drop('RowNumber', axis=1)
```

```python
df.drop(['RowNumber','CustomerId',	'Surname'	,'Geography','Gender'], axis=1,inplace=True)
```

```python
df
```

```python
# to check for null value in the dataset
df.isnull().sum()
```

```python
# split the data in 'x'
input_data = df.iloc[:,:-1]
```

```python
input_data
```

```python
output_data = df['EstimatedSalary']
output_data
```

```python
# scaling of the input_data
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
input_data= pd.DataFrame(scale.fit_transform(input_data),columns=input_data.columns)
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
input_data.shape
```

```python
ann.add(Dense(6,input_dim = 8, activation="relu"))
ann.add(Dense(4, activation="relu"))
ann.add(Dense(2, activation="relu"))
ann.add(Dense(1, activation="sigmoid"))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```
