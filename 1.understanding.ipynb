Forward Propogation and Backward Propogation

- Foward Propogation
Forward Propagation is one of the core steps in a neural network.
In this step, the input data moves layer by layer through the network, gets multiplied with weights, added with bias, and passed through activation functions to produce the final output (prediction).
Our goal in forward propagation is to calculate the predicted output based on current weights and biases.

In simple terms:

Forward propagation = “Making a prediction using current parameters.”

- Backward Propogation
Backward Propagation is used to improve the model.
After we compare the predicted output with the actual output, we calculate the error (loss). Then we move backward through the network to update the weights and bias in order to reduce that error.
Our task is to classify the data as accurately as possible, and for that we continuously update the weights and bias using the error.

In simple terms:

Backward propagation = “Learning from mistakes and adjusting parameters.”
